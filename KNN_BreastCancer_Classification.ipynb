```python
def euclidean_distance(v1,v2):
    v1,v2 = np.array(v1),np.array(v2)
    distance = 0
    for i in range(len(v1)):
        distance += (v1[i]-v2[i])**2
    return np.sqrt(distance)
```

```python
def measures(yhat,test):
    
    TP , TN , FP , FN = 0.0001 , 0.0001 , 0.0001 , 0.0001
    
    
    for ypred, y in zip(yhat, test):
        if ypred==1 and y==1:
            TP+=1
        elif ypred==1 and y==0:
            FP+=1
        elif ypred==0 and y==1:
            FN+=1
        elif ypred==0 and y==0:
            TN+=1
                
    Accuracy = (TP + TN)/len(yhat)
    
    Precision = TP / (TP + FP)
    
    Recall = TP / (TP + FN )
    
    f_measure = (2*Precision*Recall)/(Precision + Recall)
    
    print("Accuracy :",Accuracy)
    print("Precision :",Precision)
    print("Recall :",Recall)
    print("F-Measure :",f_measure)
    return f_measure
```

```python
def knnAlgo(k,train_x,train_y,test_x,test_y):
    i = 0
    j = 0
    ypred = []
    train_x = list(train_x.values)
    train_y = list(train_y)
    test_x=list(test_x.values)
    test_y =list(test_y)
   
    for i in range(len(test_x)):
        c1 = 0
        c0 = 0
        distance = []
        for j in range(len(train_x)):
            di = euclidean_distance(test_x[i],train_x[j])
            distance.append((train_x[j],di,train_y[j]))
            
        
        distance.sort(key=lambda x:x[1])
        N = distance[:k]
      
    
        for l in range(k):
            if N[l][2]==1:
                c1 = c1+1 
            if N[l][2]==0:
                c0 = c0+1 
            
        if c0>c1:
            ypred.append(0)
        else:
            ypred.append(1)  
        
    #print (ypred)
    f = measures(ypred,test_y)
    return f
```

```python
import numpy as np
from sklearn.datasets import load_breast_cancer

# Load the breast cancer dataset
breast_cancer = load_breast_cancer()

# Split the dataset into X and y
X = breast_cancer.data
y = breast_cancer.target

# Define the number of folds
k = 5

# Shuffle the indices of the samples
indices = np.random.permutation(X.shape[0])

# Split the indices into k folds
folds = np.array_split(indices, k)

# Initialize an array to store the accuracies of each fold
accuracies = np.zeros(k)

# Perform k-fold cross-validation
for i in range(k):
    # Split the data into training and testing sets for this fold
    test_indices = folds[i]
    train_indices = np.concatenate([fold for j, fold in enumerate(folds) if j != i])
    X_train, y_train = X[train_indices], y[train_indices]
    X_test, y_test = X[test_indices], y[test_indices]
    
    # Train the KNN model
    k = 5 # set the value of k
    distances = np.sqrt(np.sum((X_train[:, np.newaxis, :] - X_test[np.newaxis, :, :]) ** 2, axis=2))
    sorted_indices = np.argsort(distances, axis=0)
    top_k_indices = sorted_indices[:k]
    top_k_labels = y_train[top_k_indices]
    predicted_labels = np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=0, arr=top_k_labels)
    
    # Calculate the accuracy of the model on the test set for this fold
    accuracy = np.mean(predicted_labels == y_test)
    accuracies[i] = accuracy

# Calculate the average accuracy across all folds
avg_accuracy = np.mean(accuracies)
print("Average accuracy: {:.2f}".format(avg_accuracy))
```
